## ***Matplotlib***
Matplotlib is one of the oldest and most widely used Python visualization libraries. It provides a flexible and powerful framework for creating static, animated, and interactive plots. Its pyplot module offers MATLAB-like plotting functionality, making it a great choice for basic visualizations and fine-tuned control over figure elements.

```python
import matplotlib.pyplot as plt
import numpy as np
```

# ***Line Plot***
A line plot is a graph that displays data points connected by a line, with each data point plotted on the x-axis and y-axis. It helps in understanding trends, patterns, and changes in data, such as showing growth or decline over time.


```python
x = [1, 2, 3, 4 , 5 , 6 , 7]
y = [10, 20, 25, 30 , 75, 30 ,45]
plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

# ***Bar Chart***
A bar chart is a graphical representation of data where rectangular bars are used to represent the value of different categories.Each bar's length or height corresponds to the value it represents, and the categories are typically shown on the x-axis, while the values are on the y-axis. It is useful for comparing different groups or categories.

```python
categories = ['A', 'B', 'C','D','E','F']
values = [10, 28, 19, 67, 88, 43]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

# ***Histogram***
A histogram is a type of bar chart that represents the distribution of a dataset.It groups data into bins or intervals and the height of each bar shows how many data points fall within each bin, helping to visualize the frequency distribution of continuous data.

```python
plt.hist([1, 2, 2, 3, 4, 5, 5, 5], bins=5)
plt.show()
```

# ***Scatter Plot***
A scatter plot is a type of graph used to display the relationship between two continuous variables. It uses dots to represent data points on a two-dimensional grid, where one variable is plotted on the x-axis and the other on the y-axis. This helps in identifying correlations or patterns between the variables.

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [10, 20, 25, 39, 24, 80, 41, 70, 55]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

# ***Pie Chart***
A pie chart is a circular graph divided into slices to illustrate numerical proportions. Each slice represents a category's contribution to the whole, and the size of each slice is proportional to the category's percentage or value in the total dataset. It's useful for showing how parts make up a whole.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]
plt.pie(sizes, labels=labels)
plt.title("Pie Chart ")
plt.show()
```

# ***Box Plot***
A box plot (or box-and-whisker plot) is a graph that shows the distribution of a dataset based on five key summary values: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

It helps visualize data spread, central tendency, and potential outliers by showing the box (middle 50% of data) and whiskers (range outside the box).

```python
data = [10, 20, 25, 30, 35, 40, 45, 50, 55, 60]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
```

# ***Seaborn***
Seaborn is a Python data visualization library built on top of Matplotlib, designed to make beautiful and informative statistical graphics easily.
It provides high-level functions for creating complex plots like heatmaps, box plots, violin plots, and pair plots with just a few lines of code. Seaborn also integrates smoothly with pandas DataFrames, making it great for visualizing structured datasets.
Seaborn automatically handles themes, color palettes, and statistical summaries, making plots more attractive and insightful.
Its especially useful for visualizing relationships between multiple variables and spotting trends or patterns in data.



# **Line** **Plot**
A Seaborn line plot visually represents the relationship or trend between two continuous variables.It helps reveal patterns, trends, or fluctuations in data, often used for time series or sequential data analysis.

```python
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
sns.lineplot(x=x, y=y)
plt.title("Seaborn Line Plot")
plt.show()
```

# ***Bar Plot***
A Seaborn bar plot is used to visualize the relationship between a categorical variable and a continuous variable.
It displays rectangular bars where the length of each bar represents the value of the continuous variable, making it easy to compare different categories.

```python
categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
values = [10, 20, 15, 25, 30, 40, 35, 50]
sns.barplot(x=categories, y=values)
plt.title("Seaborn Bar Plot with 8 Values")
plt.show()
```

# ***Count Plot***
A **Seaborn count plot** is used to display the frequency of different categories in a categorical variable.
It shows the number of occurrences of each category with bars, making it useful for visualizing the distribution of categorical data.


```python
data = ['A', 'B', 'A', 'C', 'B', 'A', 'D', 'C']
sns.countplot(x=data)
plt.show()
```

# ***Histogram***
A histogram is a graphical representation of the distribution of a dataset. It divides the data into bins (intervals) and displays the frequency of data points that fall within each bin. The height of each bar represents the number of data points in that interval. It is useful for visualizing the distribution, spread, and shape of the data.

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5, 6, 7, 8]
sns.histplot(data, bins=5, kde=True)
plt.show()
```

# **Box Plot**
A box plot (or box-and-whisker plot) is used to visualize the distribution of a dataset. It shows the minimum, first quartile (Q1), median, third quartile (Q3), and maximum values of the data. The box represents the interquartile range (IQR), and the "whiskers" extend to the minimum and maximum values within a defined range. Outliers, if any, are plotted as individual points.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 7, 8, 9, 10]
sns.boxplot(data=data)
plt.show()
```

#***Pair Plot***
A pair plot is a Seaborn visualization that creates pairwise relationships between all numerical variables in a dataset. It displays scatterplots for each pair of variables along with histograms or kernel density estimates (KDE) along the diagonal. This is useful for exploring correlations and relationships between multiple variables in a dataset.

```python
data = sns.load_dataset("iris")
sns.pairplot(data)
plt.show()
```

# Comparison betwwen Matplotlib and seaborn
1. **Ease of Use**

Matplotlib:

Strength: Matplotlib provides a great level of control, allowing users to fully customize their plots. For users familiar with the library, it offers flexibility and power to create almost any kind of plot.

Weakness: It can be complex and time-consuming for beginners to create aesthetically pleasing or sophisticated plots. A lot of manual configuration is required for simple tasks (e.g., axis labels, colors, legends).

Seaborn:

Strength: Seaborn is easier to use, especially for statistical plots. It offers high-level functions that automatically handle many plot configurations like color schemes, axes, and gridlines, making it a great choice for quick and visually attractive plots.

Weakness: Seaborns high-level abstraction can sometimes limit the level of customization compared to Matplotlib, requiring more effort if users need precise control.

2. **Customization Options**

Matplotlib:

Strength: Matplotlib excels in customization. Every plot element, from titles to axis ticks, can be controlled individually, offering a very granular level of customization.

Weakness: Customization can be tedious and time-consuming. Achieving an attractive, professional-looking plot may require significant manual effort and additional lines of code.

Seaborn:

Strength: While Seaborn has a more limited set of customization options compared to Matplotlib, it provides enough flexibility to tweak most common attributes. It automatically applies aesthetic defaults that make the plots look good out of the box.

Weakness: If deep customization is required (e.g., specific control over the color of individual elements or fine-tuning the plot design), Seaborn can become less flexible and might require resorting to Matplotlib for further adjustments.

3. **Interactivity**

Matplotlib:

Strength: Matplotlib provides basic interactivity through tools like zooming and panning when used in environments like Jupyter Notebooks. It also integrates with other libraries like mplcursors and Plotly to enhance interactivity.

Weakness: It is not specifically designed for interactive plots. While some interactivity can be added, it requires additional libraries or manual configuration, making it less convenient for interactive visualizations.

Seaborn:

Strength: Seaborn is built on top of Matplotlib, so it can leverage Matplotlib's interactivity features. However, it does not provide as much built-in interactivity as specialized libraries like Plotly or Bokeh.

Weakness: Seaborn is not designed with interactivity as a primary focus. Interactive features are limited compared to other visualization libraries, and users often need to combine Seaborn with other tools like Plotly for more dynamic plots.

4.** Performance with Large Datasets**

Matplotlib:

Strength: Matplotlib can handle large datasets effectively because of its lower-level design. It offers better control over rendering, and with the right optimizations, it can scale well for complex and large datasets.

Weakness: As the dataset size increases, Matplotlib may face performance bottlenecks unless users implement specific optimizations or use other libraries to handle the data more efficiently.

Seaborn:

Strength: Seaborn is built on top of Matplotlib, so it can handle datasets of moderate size without issue. Its integration with pandas makes it convenient to work with structured data, and it performs well for typical data science tasks.

Weakness: Seaborn can struggle with very large datasets (millions of points), especially when plotting complex statistical visualizations like heatmaps or pair plots. For larger datasets, Matplotlib or other performance-oriented libraries like Datashader or Plotly might be more suitable.
